```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load a model
model = YOLO('C:\\Users\\jungf\\runs\\detect\\train3\\weights\\best.pt')  # pretrained YOLOv8n model

model.predict(source="0", show=True, stream=True, classes=[0, 1])  # [0, 3, 5] for multiple classes

# Define image file path
image_folder = 'C:\\DEV\\YOLOv8Pro\\images\\'

# Define image file names
image_file_1 = 'TEST_000.png'
image_file_2 = 'TEST_017.png'

# Construct full image file paths
image_path_1 = image_folder + image_file_1 
image_path_2 = image_folder + image_file_2
 
# Run batched inference on a list of images
results = model([image_path_1, image_path_2])  # return a list of Results objects
# model.predict(source="0", show=True, stream=True, classes=0)

# Process results list
for result in results:
    boxes = result.boxes # Boxes object for bbox outputs
    masks = result.masks # Masks object for segmentation masks outputs
    keypoints = result.keypoints # Keypoints object for pose outputs
    probs = result.probs # Class probabilities for classification outputs
    
    # Print box coordinates
    print("Box coordinates:")
    for box in boxes:
        print(box)
    # Display result with boxes
    result_image = result.plot()
    result_image
    # result.plot()
    
    cv2.waitKey(0) # destroyAllWindows 함수에 적용
    cv2.destroyAllWindows() # wait() 파라미터 키 입력 시 모든 결과창 닫기
```
